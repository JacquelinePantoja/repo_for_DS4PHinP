2. Write a class mydata that takes in a list of numbers and has the methods: mean and std (for standard deviation) returning the mean and standard deviation respectively. This should calculate the mean and standard deviation using only base python functions.
Try it out with x = mydata([1, 2, 7, 5]).

```python
#wrote the base of this code myself and applied GEMINI corrections
class mydata: #takes in a list
  def __init__(self, input_list):
    self.my_list = input_list

  def mean(self): #calculates the mean of the list
    return sum(self.my_list) / len(self.my_list)

  def std(self): #calculates the sd of the list
    n = len(self.my_list)
    data_mean = self.mean()
    variance = sum((x - data_mean) ** 2 for x in self.my_list) / (n - 1) # Sample standard deviation
    return variance ** 0.5

  def print(self): #new method that prints both the mean and sd
    return(self.mean(), self.std())

x = mydata([1, 2, 7, 5])
x.print()
```

3. Create a function called lag as follows. It takes in a numpy vector, say a = numpy.array( [1, 6, 8, 4] ) and an integer > 0. It then returns the array lagged by the integer value. So, lag(a, 1) will return [nan, 1, 6, 8] and lag(a, 2) will return [nan, nan, 1, 6]. That is, it gets rid of n values at the end and appends that many nans at the beginning. Use numpy's nan for the missing values. It should return a numpy array. Do not use numpy's existing lag, lead, roll etcetera functions. Instead, use numpy indexing.

```python
#my original code was completely off, GEMINI hated it a lot
#i applied the chnages but kept my original sytax of creating an empty array and adding nans*integer
import numpy as np

def lag(array, num_lags):
  #create an empty array and add "nan"*num_lags
  nans = np.array([])
  nans = np.append(nans, [np.nan] * num_lags)

  #wanted values are kept by removing the indexed values given by -num_lags
  kept_values = array[:-num_lags]

  # Concatenate the NaNs with the remaining part of the array
  return np.concatenate((nans, kept_values))

a = np.array( [1, 6, 8, 4] )
print(lag(a, 2))
print(lag(a, 1))
```

5. Create a function called filter_and_square. This function should take a list of numbers as input. Using a list comprehension, return a new list that contains the squares of the numbers, but only for the numbers in the original list that are greater than 0. For example, [-2, 0, 3, 4] should return [9, 16].

```python
def filter_and_square(input_list):
  new = [x for x in input_list if x > 0] #new list that only includes filtered values of og list
  return [new**2 for new in new] #squares the individual values in the list

print(filter_and_square([-2, 0, 3, 4, 8, 0, 16, -4]))
filter_and_square([-2, 0, 3, 4])
```

6. Create a function called frequency_map. This function takes a list of strings (e.g., ['apple', 'banana', 'apple', 'orange']) and returns a dictionary where the keys are the strings and the values are the counts of how many times each string appears in the list. Do not use collections.Counter; write the logic using basic dictionary operations.

```python
def frequency_map(input_list):
  frequency_map = {x: input_list.count(x) for x in set(input_list)} #set() returns list of unique items in list, and .count() returns the count of that unique item in a list
  return frequency_map

list = ['apple', 'banana', 'apple', 'orange', 'orange', 'watermelon', "pear", 'pear', 'pear']
frequency_map(list)
```
